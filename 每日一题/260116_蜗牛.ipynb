这一道题目是以一个蜗牛爬行为基础，是一个动态规划的题目

所以我们可以维护两个状态
一个是到达目前的底部的最短时间
一个是到达目的杆子上的最短时间

f[i]：到达 (x_i, 0) 的最短时间
g[i]：到达 (x_i, A[i]) 的最短时间（只有 i < n 时有用，因为要从这里传送）

状态转移方程
dp[i+1]=dp[i]+min(xi+1-xi,ai/0.7+bi+1/1.3)



```python
n = int(input())
x = list(map(int, input().split()))

if n == 1:
    ans = x[0]
    print(f"{ans:.2f}")
else:
    a_list = []
    b_list = []

    for i in range(n-1):
        a_val, b_val = map(int, input().split())
        a_list.append(a_val)
        b_list.append(b_val)

    dp = x[0]
    for i in range(n-1):
        walk_time = x[i+1] - x[i]
        teleport_time = a_list[i]/0.7 + b_list[i]/1.3
        dp = dp + min(walk_time, teleport_time)

    print(f"{dp:.2f}")


```
